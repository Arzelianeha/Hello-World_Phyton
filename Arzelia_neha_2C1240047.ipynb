```python
jumlah_barang = int(input("jumlah barang :"))
total_belanja = 0
for i in range(jumlah_barang):
  harga = int(input(f"harga barang ke-{i+1}: "))
  jumlah = float(input("jumlah unit :"))
  total_barang = harga * jumlah
  print("total harga barang adalah", total_barang)
  total_belanja += total_barang
print ("Total belanja =", total_belanja)
```

```python
# iterasi pada list
skills = ["Python", "R", "Julia", "Scala"]
for skill in skills:
    print("Anda menguasai", skill)
print("-------------------------")
# iterasi pada list dengan menyertakan indeks (key)
for idx, skill in enumerate(skills):
    print(f"Skill ke-{idx+1} : {skill}")
```

```python
# Program Python dengan Nested Loop
# Input jumlah baris
baris = int(input("Masukkan jumlah baris: "))
# Loop untuk membuat pola
for i in range(1, baris + 1):
    for j in range(1, i + 1):
        print("*", end=" ")
    print()  # Pindah ke baris berikutnya
```
